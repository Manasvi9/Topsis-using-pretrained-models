```python
!pip install transformers datasets evaluate torch numpy pandas matplotlib rouge_score

import torch
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM
from datasets import load_dataset
import evaluate

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

dataset = load_dataset("cnn_dailymail", "3.0.0", split="test[:20]")
bleu_metric = evaluate.load("bleu")
rouge_metric = evaluate.load("rouge")


models_to_test = [
    "gpt2",
    "google/flan-t5-small",
    "facebook/bart-base"
]

results = []

for model_name in models_to_test:
    print(f"\nEvaluating {model_name}...")

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    if "t5" in model_name or "bart" in model_name:
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)
    else:
        model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

    model.eval()

    predictions = []
    references = []
    total_time = 0
    total_loss = 0

    for sample in dataset:
        input_text = sample["article"][:512]
        reference = sample["highlights"]
        if "t5" in model_name:
            input_text = "summarize: " + input_text

        inputs = tokenizer(input_text, return_tensors="pt", truncation=True).to(device)
        start_time = time.time()

        with torch.no_grad():
            output = model.generate(
                **inputs,
                max_new_tokens=100,
                do_sample=False
            )

        end_time = time.time()
        total_time += (end_time - start_time)

        generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

        predictions.append(generated_text)
        references.append(reference)
        with torch.no_grad():
            if "t5" not in model_name and "bart" not in model_name:
                outputs = model(**inputs, labels=inputs["input_ids"])
                loss = outputs.loss
                total_loss += loss.item()
            else:
                total_loss += 0

    bleu = bleu_metric.compute(predictions=predictions, references=[[r] for r in references])["bleu"]
    rouge = rouge_metric.compute(predictions=predictions, references=references)["rougeL"]
    if len(dataset) > 0 and total_loss > 0:
        perplexity = np.exp(total_loss / len(dataset))
    else:
        perplexity = np.nan

    avg_time = total_time / len(dataset)
    model_size = sum(p.numel() for p in model.parameters()) / 1e6

    results.append([
        model_name,
        bleu,
        rouge,
        perplexity,
        avg_time,
        model_size
    ])


columns = ["Model", "BLEU", "ROUGE", "Perplexity", "Time", "Size"]
df = pd.DataFrame(results, columns=columns).set_index("Model")

print("\nEvaluation Results:\n")
print(df)

def topsis(data, weights, criteria):
    data = data.values
    norm = data / np.sqrt((data ** 2).sum(axis=0))
    weighted = norm * weights

    ideal_best = np.max(weighted * criteria + weighted * (1 - criteria) * -1, axis=0)
    ideal_worst = np.min(weighted * criteria + weighted * (1 - criteria) * -1, axis=0)

    s_plus = np.sqrt(((weighted - ideal_best) ** 2).sum(axis=1))
    s_minus = np.sqrt(((weighted - ideal_worst) ** 2).sum(axis=1))

    scores = s_minus / (s_plus + s_minus)
    return scores
weights = np.array([0.25, 0.25, 0.20, 0.15, 0.15])

criteria = np.array([1, 1, 0, 0, 0])
max_perplexity = df['Perplexity'].max()
if pd.isna(max_perplexity):
    fill_value = 1e9
else:
    fill_value = max_perplexity * 1.1 + 100

df_for_topsis = df.copy()
df_for_topsis['Perplexity'] = df_for_topsis['Perplexity'].fillna(fill_value)

scores = topsis(df_for_topsis, weights, criteria)

df["TOPSIS Score"] = scores
df["Rank"] = df["TOPSIS Score"].rank(ascending=False)

df = df.sort_values("TOPSIS Score", ascending=False)

print("\nFinal Ranking:\n")
print(df)

plt.figure()
plt.bar(df.index, df["TOPSIS Score"])
plt.xlabel("Models")
plt.ylabel("TOPSIS Score")
plt.title("TOPSIS Ranking (Text Generation)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
